```r
a_to_b <- function(a, b) {
    # a_to_b(1, 1)
    # [1] 1
    # a_to_b(1, 3)
    # [1] 1 2 3
    # a_to_b(1, 5)
    # [1] 1 2 3 4 5
    ### BEGIN SOLUTION
    return(a:b)
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `find_positives`ï¼Œçµ¦å®šä¸€å€‹æ•¸å€¼å‘é‡èƒ½å¤ å›å‚³å…¶ä¸­çš„æ­£æ•¸ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡

```r
find_positives <- function(x) {
    # find_positives(-3:3)
    # [1] 0 1 2 3
    # find_positives(c(-1, 1, -2, 2, -3, 3))
    # [1] 1 2 3
    ### BEGIN SOLUTION
    return(x[x >= 0])
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `find_odds`ï¼Œçµ¦å®šä¸€å€‹æ•¸å€¼å‘é‡èƒ½å¤ å›å‚³å…¶ä¸­çš„å¥‡æ•¸ã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡

```r
find_odds <- function(x) {
    # find_odds(1:3)
    # [1] 1 3
    # find_odds(10:15)
    # [1] 11 13 15
    # find_odds(2:6)
    # [1] 3 5
    ### BEGIN SOLUTION
    return(x[x %% 2 == 1])
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `find_evens`ï¼Œçµ¦å®šèµ·é» a åˆ°çµ‚é» b ä¹‹é–“çš„æ•¸å€¼å‘é‡ï¼Œå…¶ä¸­ $a \leq b$ï¼Œå›å‚³å…¶ä¸­çš„å¶æ•¸ã€‚

- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡

```r
find_evens <- function(a, b) {
    # find_evens(1, 3)
    # 2
    # find_evens(10:15)
    # [1] 10 12 14
    # find_evens(2:6)
    # [1] 2 4 6
    ### BEGIN SOLUTION
    x <- a:b
    return(x[x %% 2 == 0])
    ### END SOLUTION
}
```
