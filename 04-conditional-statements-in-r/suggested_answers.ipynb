## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `calculate_circle` èƒ½å¤ å°‡è¼¸å…¥çš„åŠå¾‘ä¾ç…§åƒæ•¸ `to` è¨ˆç®—ç‚ºé¢ç©æˆ–å‘¨é•·ï¼Œä½¿ç”¨ $\pi = 3.14$ã€‚

\begin{equation}
Area = \pi r^2 \\
Perimeter = 2 \pi r \\
\text{where} \; \pi = 3.14
\end{equation}

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€ä¸€å€‹æ–‡å­—å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ–‡å­—å‘é‡

```r
calculate_circle <- function(r, to) {
    # calculate_circle(3, to='area')
    # 28.26
    # calculate_circle(3, to='perimeter')
    # 18.84
    # calculate_circle(4, to='area')
    # 50.24
    # calculate_circle(4, to='perimeter')
    # 25.12
    ### BEGIN SOLUTION
    my_pi <- 3.14
    if (to == 'area') {
        return(my_pi*r**2)
    } else {
        return(2*my_pi*r)
    }
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `convert_degree` èƒ½å¤ å°‡è¼¸å…¥çš„æº«åº¦åº¦æ•¸ä¾ç…§åƒæ•¸ `from` èˆ‡ `to` åœ¨æ”æ°èˆ‡è¯æ°ä¹‹é–“è½‰æ›ã€‚

\begin{equation}
Fahrenheit_{(Â°F)} = Celsius_{(Â°C)} \times \frac{9}{5} + 32 \\
Celsius_{(Â°C)} =  (Fahrenheit_{(Â°F)} - 32) \times \frac{5}{9}
\end{equation}

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡ã€å…©å€‹æ–‡å­—å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ•¸å€¼å‘é‡

```r
convert_degree <- function(x, from, to) {
    # convert_degree(100, from='Celsius', to='Fahrenheit')
    # 212.0
    # convert_degree(0, from='Celsius', to='Fahrenheit')
    # 32.0
    # convert_degree(212, from='Fahrenheit', to='Celsius')
    # 100.0
    # convert_degree(32, from='Fahrenheit', to='Celsius')
    # 0.0
    ### BEGIN SOLUTION
    if (from == 'Celsius') {
        return(x*9/5 + 32)
    } else {
        return((x - 32)*5/9)
    }
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `get_bmi_category` èƒ½å¤ å°‡è¼¸å…¥çš„èº«é«˜ï¼ˆå…¬åˆ†ï¼‰ã€é«”é‡ï¼ˆå…¬æ–¤ï¼‰æ›ç®—ç‚º BMI é¡åˆ¥ã€‚

\begin{equation}
BMI = \frac{weight_{kg}}{height_{m}^2}
\end{equation}

<https://en.wikipedia.org/wiki/Body_mass_index>

- é æœŸè¼¸å…¥ï¼šå…©å€‹æ•¸å€¼å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ–‡å­—å‘é‡

```r
get_bmi_category <- function(height, weight) {
    # get_bmi_category(170, 50)
    # 'Underweight'
    # get_bmi_category(170, 60)
    # 'Normal'
    # get_bmi_category(206, 113)
    # 'Overweight'
    # get_bmi_category(216, 147)
    # 'Obese'
    ### BEGIN SOLUTION
    bmi <- weight / (height*0.01)**2
    if (bmi > 30) {
        category <- 'Obese'
    } else if (bmi >= 25) {
        category <- 'Overweight'
    } else if (bmi >= 18.5) {
        category <- 'Normal'
    } else {
        category <- 'Underweight'
    }
    return(category)
    ### END SOLUTION
}
```

```r
fizz <- function(x) {
    # fizz(3)
    # 'Fizz'
    # fizz(6)
    # 'Fizz'
    # fizz(7)
    # 7
    # fizz(8)
    # 8
    ### BEGIN SOLUTION
    if (x %% 3 == 0) {
        return('Fizz')
    } else {
        return(x)
    }
    ### END SOLUTION
}
```

## è‡ªè¨‚ä¸€å€‹å‡½æ•¸ `fizz_buzz` ä¾æ“šæ˜¯å¦ç‚º 3ã€5 æˆ– 15 çš„å€æ•¸å›å‚³å°æ‡‰æ–‡å­—å‘é‡ï¼Œå¦‚æœæ˜¯ 3 çš„å€æ•¸è¼¸å‡º `"Fizz"`ï¼Œå¦‚æœæ˜¯ 5 çš„å€æ•¸è¼¸å‡º `"Buzz"`ï¼Œå¦‚æœæ˜¯ 15 çš„å€æ•¸è¼¸å‡º `"Fizz Buzz"`ï¼Œå¦å‰‡å°‡è¼¸å…¥çš„æ•¸å­—ç›´æ¥è¼¸å‡ºã€‚

- é æœŸè¼¸å…¥ï¼šä¸€å€‹æ•¸å€¼å‘é‡
- é æœŸè¼¸å‡ºï¼šä¸€å€‹æ–‡å­—å‘é‡æˆ–ä¸€å€‹æ•¸å€¼å‘é‡

```r
fizz_buzz <- function(x) {
    # fizz_buzz(3)
    # 'Fizz'
    # fizz_buzz(5)
    # 'Buzz'
    # fizz_buzz(15)
    # 'Fizz Buzz'
    # fizz_buzz(16)
    # 16
    ### BEGIN SOLUTION
    if (x %% 15 == 0) {
        return('Fizz Buzz')
    } else if (x %% 3 == 0) {
        return('Fizz')
    } else if (x %% 5 == 0) {
        return('Buzz')
    } else {
        return(x)
    }
    ### END SOLUTION
}
```
